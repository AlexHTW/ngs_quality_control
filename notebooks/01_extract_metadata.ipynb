## Get metadata from path and filename and export to file:
* path
* organism: E. faecium (Efcm), E.coli(Ecoli), S. aureus (Sau)
* sequencing technology: HS, MS, NX
* Read number: 1/2
* evaluation (label): good/ugly (1/0)

```python
import os
import re
import pandas as pd
```

```python
rootdir = 'training_data'
```

```python
reads =	{
  "filename": [],
  "organism": [],
  "technology": [],
  "read_number": [],
  "evaluation": []
}
```

```python
for root, dirs, files in os.walk(rootdir):
    if root.endswith("fastqc"):
        extracted_vars = root.split('/')
        reads["organism"].append(extracted_vars[-5])
        reads["evaluation"].append(extracted_vars[-4])
        reads["technology"].append(extracted_vars[-3])
        reads["filename"].append(extracted_vars[-1].replace("_fastqc", ""))
        read_number = extracted_vars[-1].split('_')[-3][1]
        reads["read_number"].append(read_number)
```

```python
metadata = pd.DataFrame(reads)
metadata
```

```python
metadata.to_json('exported_datasets/metadata.json')
```
