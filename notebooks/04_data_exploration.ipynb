#### variable data size betweend reads

Tables within a module (column) have different lengths, which is problematic for usage as features in ML.  
Show different table lengths of all modules.  
Select unique combinations of lengths, organism and technology to check for correlation.

```python
def unique_size_combinations(df, module_name):
    all_rows = []
    for i in range(df.shape[0]):
        count = str(df[module_name][i].shape[0]) + ' ' + df['organism'][i] + ' ' + df['technology'][i]
        all_rows.append(count)
    unique_combinations = Counter(all_rows).keys()
    combination_counts = Counter(all_rows).values()
    unique_counts = pd.DataFrame()
    unique_counts['combinations'] = unique_combinations
    unique_counts['counts'] = combination_counts
    return unique_counts
```

```python
counts_per_base_sequence_quality = unique_size_combinations(ngs_reads, 'Per base sequence quality')
counts_per_tile_sequence_quality = unique_size_combinations(ngs_reads, 'Per base sequence quality')
counts_per_sequence_quality_scores = unique_size_combinations(ngs_reads, 'Per sequence quality scores')
counts_per_base_sequence_content = unique_size_combinations(ngs_reads, 'Per base sequence content')
counts_per_sequence_gc_content = unique_size_combinations(ngs_reads, 'Per sequence GC content')
counts_per_base_n_content = unique_size_combinations(ngs_reads, 'Per base N content')
counts_sequence_length_distribution = unique_size_combinations(ngs_reads, 'Per base sequence quality')
counts_sequence_duplication_levels = unique_size_combinations(ngs_reads, 'Sequence Duplication Levels')
counts_overrepresented_sequences = unique_size_combinations(ngs_reads, 'Overrepresented sequences')
counts_adapter_content = unique_size_combinations(ngs_reads, 'Adapter Content')
```
